# 线性回归模型

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
""" 数据准备阶段 """
# x_train：训练数据的特征矩阵，通常是一个二维数组，每一行代表一个样本，每一列代表一个特征。
# y_train：训练数据的目标值向量，是一个一维数组，与 x_train 中的样本一一对应。
x_data = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1)
y_data = np.array([1, 2, 3, 4, 5, 5, 6, 7]).reshape(-1, 1)
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2)
```

```python
""" 模型创建 """
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
""" 模型评估 """
score = model.score(x_test, y_test)
print(f"模型的评估结果为：{score}")
```

```python
""" 模型预测 """
y_predict = model.predict(x_test)
print(f"预测结果为：{y_predict}")
```

```python
""" 可视化 """
plt.scatter(x_test, y_test, color='red')
plt.plot(x_test, y_predict, color='blue')
plt.show()
```
